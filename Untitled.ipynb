```python
#libraries
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# figure parametres
plt.rcParams['figure.figsize'] = [11, 4]
plt.rcParams['figure.dpi'] = 100
```

```python
data=pd.read_csv('netflix_titles.csv')
```

```python
data
```

## Task 1: Count the number of shows added per day     


```python
# convert date from object to date
```

```python
data.dtypes
```

```python
data['date_added']=pd.to_datetime(data['date_added'])
```

```python
data.dtypes
```

```python
# count show added per date
df=data.copy()
```

```python
show_added=df.groupby(df['date_added'])['show_id'].count()
```

```python
show_added
```

```python
show_added.describe()
```

```python
# rename the columns
show_added = show_added.rename({'show_id': 
                                  'number_of_shows_added'}, 
                                 axis=1)
```

```python
show_added
```

```python
show_added.plot()
```

## Task 2: Extract the day name and sum-up the shows added 


```python
df['day']=df['date_added'].dt.day_name()
```

```python
show_added_day=df.groupby(df['day'])['show_id'].count()
```

```python
show_added_day
```

# Task 3: Select data from 2016 onwards

```python
show=df[df['date_added'] >= '1-1-2016']
```

```python
show_2016=show.groupby(df['date_added'])['show_id'].count()
```

```python
show_2016
```

```python
show_2016.plot()
```

```python
# Use the resampling function to group data per week
weekly_data = (show_2016.
               resample('1W')   # For each week 
               .sum())  
```

```python
weekly_data.plot()
```
